```python
nlp = en_core_web_sm.load()
```

```python
hash_df = {}

text = ""
for index, row in rdata_json.iterrows():
    # if (index % 100 == 0):
        # print("index: ", index)
    sample = row["content"]
    text = text + sample
len(text)
```

```python
text_list = list(map(''.join, zip(*[iter(text)]*100000)))
text_list[0]
```

```python
for i, text in enumerate(text_list):
    if (i % 1000 == 0):
        print("index: ", i)
    doc = nlp(text)
    # Analyze syntax
    # print("Noun phrases:", [chunk.text for chunk in doc.noun_chunks])
    # print("Verbs:", [token.lemma_ for token in doc if token.pos_ == "VERB"])
    # Find named entities, phrases and concepts
    # entity_label = pd.DataFrame(columns=['label'])
    for entity in doc.ents:
        # print(entity.text, entity.label_)
        if entity.label_ not in hash_df:
            hash_df[entity.label_] = 0
        hash_df[entity.label_] = hash_df[entity.label_] + 1
```
